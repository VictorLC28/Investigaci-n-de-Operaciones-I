```python
import sympy as sp
sp.init_printing()
```

```python
x=sp.symbols('x')
y=sp.Function('y')
```

```python
ode=sp.Eq(y(x)*sp.Derivative(y(x),x)+2*x,0)
```

```python
ode
```

```python
sp.classify_ode(ode)
```

```python
sol=sp.dsolve(ode,y(x))
```

```python
sol
```

```python
from sympy.plotting import plot3d_parametric_surface
```

```python
u,v=sp.symbols('u v')
```

```python
plot3d_parametric_surface(sp.cos(u+v),sp.sin(u-v),u-v)
```
