## Problema

Una fábrica produce dos tipos de recipientes, K y L, para lo cual utiliza dos máquinas, $M_1$ y $M_2$. Para producir un recipiente K, $M_1$ necesita 2 minutos y $M_2$, 4 minutos. De manera semejante, L ocupa 8 minutos a $M_1$ y 4 minutos a $M_2$. La ganancia neta para un recipiente K es de 29 pesos y para uno L es de 45 pesos. Determinar el plan de producción que maximice la ganancia neta.

De manera formal:

#### Variables de decisión:

Número de recipientes K, $x_1$ y L, $x_2$.

#### Función Objetivo:

Maximizar $29x_1 + 45x_2$

#### Restricciones:

Consideramos la producción por hora:

Resultante de la máquina $M_1$: $2x_1 + 8x_2 \leq 60$

Resultante de la máquina $M_2$: $4x_1 + 4x_2 \leq 60$

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

#### Ejercicio 1

Una compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor A gana $ \$  3$ por motor y con el motor B gana $\$ 4$ por motor. Dispone de 80 horas por semana para producirlos y en los procesos se requieren: 4 h para A y solo 2 h para B. Sin embargo, debido a lo peligroso del material usado, son necesarias 5 h de preparación y limpieza para el motopr B y 2 h para A. El tiempo de preparación total disponible es 120 h por semana. Determinar el número de cada clase de motores a producir que dé como resultado la máxima ganancia.

De manera formal:

Variables de decisión:
- Tipo de combustible A: $x_1$  
- Tipo de combustible B: $x_2$

Función Objetivo:

- Maximizar  $3x_1+4x_2$

Restricciones:

Consideramos la producción por hora:

- Horas por semana para producción:  $4x_1+2x_2≤80$
- Horas por semana para preparación:  $2x_1+5x_2≤120$

No negatividad:  $x_1≥0$, $x_2≥0$

```python
plt.figure()

# Horas por semana para producción.
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='blue', alpha=0.25, label='Hrs. por semana para Prod.')

plt.xlabel('Cantidad de motores tipo A')
plt.ylabel('Cantidad de motores tipo B')
plt.legend()
plt.show()
```

```python
plt.figure()

# Horas por semana para producción.
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='blue', alpha=0.25, label='Hrs. por semana para Prod.')

# Horas por semana para preparación
plt.plot([0, 60.0], [25.0, 0], color='red', alpha=0.125)
plt.fill_between([0, 60.0], [25.0, 0], color='red', alpha=0.125, label='Hrs. por semana para Prep.')

plt.xlabel('Cantidad de motores tipo A')
plt.ylabel('Cantidad de motores tipo B')
plt.legend()
plt.show()
```

```python
plt.figure()

# Horas por semana para producción.
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='blue', alpha=0.25, label='Hrs. por semana para Prod.')

# Horas por semana para preparación
plt.plot([0, 60.0], [25.0, 0], color='red', alpha=0.125)
plt.fill_between([0, 60.0], [25.0, 0], color='red', alpha=0.125, label='Hrs. por semana para Prep.')

plt.xlabel('Cantidad de motores tipo A')
plt.ylabel('Cantidad de motores tipo B')

plt.scatter([8], [18])  # punto factible

plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

Si trazamos la curva sobre el punto factible propuesto anteriormente $(8,18)$ tenemos: $$z=3(8)+4(18)=96$$
es decir, obtenemos la curva de nivel $$3x_1+4x_2=96$$
Es decir:


```python
plt.figure()

# Horas por semana para producción.
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='blue', alpha=0.25, label='Hrs. por semana para Prod.')

# Horas por semana para preparación
plt.plot([0, 60.0], [25.0, 0], color='red', alpha=0.125)
plt.fill_between([0, 60.0], [25.0, 0], color='red', alpha=0.125, label='Hrs. por semana para Prep.')

plt.xlabel('Cantidad de motores tipo A')
plt.ylabel('Cantidad de motores tipo B')

plt.scatter([8], [18])  # punto factible
plt.plot([0,32.0], [24.0,0], color='black')

plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```

Finalmente, notemos que geométricamente la solución de nuestro problema se da en el punto de interseción de nuestras restricciones, i.e, la solución al sistema:

$$  \left\{\begin{matrix}
 4x_1+2x_2=80\\2x_1+5x_2=120
\end{matrix}\right. $$

la cual es el punto $(10, 20)$, por tanto la curva de nivel que pasa por dicho punto es:

$$3x_1+4x_2=110$$

Gráficamente:

```python
plt.figure()

# Horas por semana para producción.
plt.plot([0, 20.0], [40.0, 0], color='blue', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='blue', alpha=0.25, label='Hrs. por semana para Prod.')

# Horas por semana para preparación
plt.plot([0, 60.0], [24.0, 0], color='red', alpha=0.125)
plt.fill_between([0, 60.0], [24.0, 0], color='red', alpha=0.125, label='Hrs. por semana para Prep.')

plt.xlabel('Cantidad de motores tipo A')
plt.ylabel('Cantidad de motores tipo B')

plt.scatter([8], [18])  # punto factible
plt.plot([0,32.0], [24.0,0], color='black', linestyle='--')

plt.scatter([10], [20], color='red')  # punto factible
plt.plot([0, 110.0/3.0], [110.0/4, 0], color='red', linestyle='-.')


plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```
