```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,6,50)
```

```python
plt.figure()
plt.plot([0,6.0],[3.0,0], color='red', alpha=0.125)
plt.fill_between([0,6.0],[3.0,0], color='red',alpha=0.125, label="Materia prima M1")

plt.plot([0,4.0],[6.0,0], color='blue', alpha=0.125)
plt.fill_between([0,4.0],[6.0,0], color='blue',alpha=0.125, label="Materia prima M2")

plt.plot([0,21.0/5.0],[21.0/4.0,0], color='black', alpha=0.35)

plt.plot(x,x+1, color='green', alpha=0.25)

plt.axhline(2)

plt.scatter([3.0],[1.5])

plt.xlabel("Pintura exteriores")
plt.ylabel("Pintura interiores")
plt.legend()

plt.xlim(0)
plt.ylim(0)
plt.show()
```
